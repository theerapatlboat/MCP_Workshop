```python
from openai import OpenAI # เรียกใช้งาน OpenAI Library
API_KEY="REDACTED_OPENAI_API_KEY" # ใช้งาน GPT-5 หรือ GPT ตัวอื่น ๆ
client = OpenAI(
    api_key = API_KEY,
)
base_model = "gpt-5.2"
```

## ตัวอย่าง

```python
res = client.chat.completions.create(
    model=base_model,
    messages=[
        {
            "role": "user",
            "content": """
            ข้าวมันไก่
            """
        },
    ],
    stream=False
)
print(res.choices[0].message.content)
```

```python
res = client.chat.completions.create(
    model=base_model,
    messages=[
        {
            "role": "user",
            "content": """
            Previous conversation : ข้าวมันไก่ทำยังไง
            ==========
            ข้าวมันไก่
            """
        },
    ],
    stream=False
)
print(res.choices[0].message.content)
```
